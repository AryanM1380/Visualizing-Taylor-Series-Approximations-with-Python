```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

```python
def taylor_exp(a, x_vals, n):
    """
    Calculates the Taylor series approximation of e^x around point a.
    
    Parameters:
        a (float): The center of the Taylor expansion
        x_vals (numpy array): Array of x values
        n (int): Number of terms to include in the series
    
    Returns:
        numpy array: Approximated values of e^x
    """
    approx = np.zeros_like(x_vals)
    for i in range(n):
        approx += (np.exp(a) * (x_vals - a)**i) / factorial(i)
    return approx
```

```python
# Define the x-axis values
x_vals = np.linspace(-2, 2, 400)

# Taylor expansion center (a = 0) and number of terms
a = 0
n = 5

# Calculate true and approximated values
true_vals = np.exp(x_vals)
taylor_vals = taylor_exp(a, x_vals, n)

# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(x_vals, true_vals, label='Original Function: $e^x$', color='blue')
plt.plot(x_vals, taylor_vals, label=f'Taylor Approximation (n={n})', color='red', linestyle='--')
plt.title('Taylor Series Approximation of $e^x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
import ipywidgets as widgets
from IPython.display import display

# Taylor series approximation function with center a
def taylor_series(func_name, a, x_vals, n):
    approx = np.zeros_like(x_vals)
    for i in range(n):
        if func_name == 'exp(x)':
            term = (np.exp(a) * (x_vals - a)**i) / factorial(i)

        elif func_name == 'sin(x)':
            if i % 2 == 0:
                term = 0
            else:
                term = ((-1)**((i - 1)//2) * (x_vals - a)**i) / factorial(i)

        elif func_name == 'cos(x)':
            if i % 2 != 0:
                term = 0
            else:
                term = ((-1)**(i//2) * (x_vals - a)**i) / factorial(i)

        elif func_name == 'ln(1 + x)':
            # limit calculations to when (x - a) > -1
            if i == 0:
                term = 0
            else:
                term = ((-1)**(i+1) * (x_vals - a)**i) / i

        elif func_name == '1 / (1 - x)':
            term = (x_vals - a)**i

        elif func_name == 'arctan(x)':
            if i % 2 == 0:
                term = 0
            else:
                term = ((-1)**((i - 1)//2) * (x_vals - a)**i) / i

        else:
            term = 0

        approx += term
    return approx

# Plotting function with center a
def plot_taylor(func_name, a, n):
    # Set x range depending on function and center point
    if func_name in ['ln(1 + x)', '1 / (1 - x)']:
        x_min = max(-0.9, a - 2)
        x_max = min(0.9, a + 2)
        x_vals = np.linspace(x_min, x_max, 400)
    else:
        x_vals = np.linspace(a - 2, a + 2, 400)

    # Actual function values
    if func_name == 'exp(x)':
        true_vals = np.exp(x_vals)
    elif func_name == 'sin(x)':
        true_vals = np.sin(x_vals)
    elif func_name == 'cos(x)':
        true_vals = np.cos(x_vals)
    elif func_name == 'ln(1 + x)':
        true_vals = np.log(1 + x_vals)
    elif func_name == '1 / (1 - x)':
        true_vals = 1 / (1 - x_vals)
    elif func_name == 'arctan(x)':
        true_vals = np.arctan(x_vals)
    else:
        true_vals = x_vals

    # Taylor series approximation values
    taylor_vals = taylor_series(func_name, a, x_vals, n)

    # Plotting
    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, true_vals, label=f'Original Function: {func_name}', color='blue')
    plt.plot(x_vals, taylor_vals, label=f'Taylor Approximation (n={n}, a={a})', color='red', linestyle='--')
    plt.title(f'Taylor Series Approximation of {func_name} around a={a}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Widgets
function_selector = widgets.Dropdown(
    options=['exp(x)', 'sin(x)', 'cos(x)', 'ln(1 + x)', '1 / (1 - x)', 'arctan(x)'],
    value='exp(x)',
    description='Function:'
)

a_slider = widgets.FloatSlider(value=0.0, min=-2.0, max=2.0, step=0.1, description='Center a:')

term_slider = widgets.IntSlider(value=5, min=1, max=20, step=1, description='n terms:')

# Interactive output
interactive_plot = widgets.interactive_output(plot_taylor, {
    'func_name': function_selector,
    'a': a_slider,
    'n': term_slider
})

# Display widgets and plot
display(function_selector, a_slider, term_slider, interactive_plot)
```
